```python
# --- 1. Trustworthiness Baseline: Calculating the Trust_Score (Weighted Rating) ---
# C: Mean rating across all restaurants
C = df['Rating'].mean()
C
```

```python
# m: Minimum votes required (using the 80th percentile as a threshold)
m = df['Votes'].quantile(0.8)
m
```

```python
def weighted_rating(df, m, C):
    R = df['Rating']
    v = df['Votes']
    # Formula: (v / (v + m) * R) + (m / (v + m) * C)
    return (v / (v + m) * R) + (m / (v + m) * C)

df['Trust_Score'] = df.apply(weighted_rating, args=(m, C), axis=1)
df
```

# Feature Engineering

```python
# --- 2. Feature Engineering for Health and Mood ---
# Prepare the Cuisine column by lowercasing and removing spaces
df['Cuisine_List'] = df['Cuisine'].str.lower().str.replace(' ', '').str.split(',')
```

```python
# Define Cuisines for Scoring (Includes the necessary 'healthfood' fix)
HEALTHY_CUISINES = ['salad', 'healthfood', 'vegan', 'smoothies', 'glutenfree', 'juices', 'healthyeats', 'health']
UNHEALTHY_CUISINES = ['desserts', 'icecream', 'fastfood', 'pizza', 'burger', 'bakery', 'streetfood']
FINE_DINING_CUISINES = ['continental', 'italian', 'seafood', 'european', 'french', 'mediterranean']
CASUAL_CUISINES = ['cafe', 'fingerfood', 'northindian', 'chinese', 'mughlai']
```

```python
# Calculate Median Cost for Mood Factor
median_cost = df['Cost'].median()
median_cost
```

```python
# a) Calculate Health Score
def calculate_health_score(cuisine_list):
    score = 0
    for cuisine in cuisine_list:
        if cuisine in HEALTHY_CUISINES:
            score += 1
        elif cuisine in UNHEALTHY_CUISINES:
            score -= 1
    return score

df['Health_Score'] = df['Cuisine_List'].apply(calculate_health_score)
```

```python

# b) Calculate Mood Score (Fine Dining vs. Casual)
def calculate_mood_score(row):
    cuisine_list = row['Cuisine_List']
    cost = row['Cost']
    score = 0
    for cuisine in cuisine_list:
        if cuisine in FINE_DINING_CUISINES:
            score += 1
        elif cuisine in CASUAL_CUISINES:
            score -= 1
    # Add Cost Factor: higher cost biases towards Fine Dining (+1)
    if cost > median_cost:
        score += 1
    elif cost < median_cost:
        score -= 1
    return score

df['Mood_Score'] = df.apply(calculate_mood_score, axis=1)
```

```python
# --- 3. Feature Engineering for Tourist Relevance ---

# Define known tourist/upscale areas (pre-processed to match the cleaned format)
TOURIST_LOCALITIES = ['centraldelhi', 'connaughtplace', 'sarabhanagar', 'sector17']

def calculate_tourist_relevance(df):
    locality = df['Locality'].str.lower().str.replace(' ', '')
    city = df['City'].str.lower().str.replace(' ', '')
    is_tourist_locality = locality.str.contains('|'.join(TOURIST_LOCALITIES), na=False) | city.str.contains('|'.join(TOURIST_LOCALITIES), na=False)
    return np.where(is_tourist_locality, 1, 0)

df['Tourist_Relevance'] = calculate_tourist_relevance(df)
```

```python
print("--- Final DataFrame Head with New Features ---")
print(df[['Name', 'Rating', 'Votes', 'Trust_Score', 'Health_Score', 'Mood_Score', 'Tourist_Relevance']].head(5).to_markdown(index=False, numalign="left", stralign="left"))
print("\n")

print("--- SCENARIO 1: Tourist-Based Recommendation (Highest Trust Score in Tourist Areas) ---")
df_tourist = df[df['Tourist_Relevance'] == 1]
tourist_recommendations = df_tourist.sort_values(by='Trust_Score', ascending=False)
print(tourist_recommendations[['Name', 'City', 'Locality', 'Trust_Score']].head(5).to_markdown(index=False, numalign="left", stralign="left"))
```

```python
# --- MASTER RECOMMENDATION FUNCTION (The Final Model) ---
def recommend_restaurant(df, user_mood="any", user_health="any", user_tourist=False, top_n=5):
    """
    Generates restaurant recommendations based on user preferences.
    Mood options: 'fine_dining', 'casual', 'any'
    Health options: 'healthy', 'indulgent', 'any'
    Tourist option: True/False
    """

    filtered_df = df.copy()

    # 1. Apply Mood Filter
    if user_mood == "fine_dining":
        filtered_df = filtered_df[filtered_df['Mood_Score'] >= 2]
    elif user_mood == "casual":
        filtered_df = filtered_df[filtered_df['Mood_Score'] <= 0]

    # 2. Apply Health Filter
    if user_health == "healthy":
        filtered_df = filtered_df[filtered_df['Health_Score'] >= 1]
    elif user_health == "indulgent":
        filtered_df = filtered_df[filtered_df['Health_Score'] <= -1]

    # 3. Apply Tourist Filter
    if user_tourist:
        filtered_df = filtered_df[filtered_df['Tourist_Relevance'] == 1]

    if filtered_df.empty:
        return "No restaurants match your combined criteria. Try broadening your search."

    # 4. Final Ranking by Trustworthiness (Trust_Score)
    recommendations = filtered_df.sort_values(by='Trust_Score', ascending=False)

    return recommendations[['Name', 'City', 'Cuisine', 'Cost', 'Trust_Score']].head(top_n)
```

```python
# --- DEMONSTRATION ---
print("--- Master Function Example: Tourist, Fine Dining, Any Health ---")
result = recommend_restaurant(df, user_mood="fine_dining", user_health="any", user_tourist=True, top_n=5)
if isinstance(result, str):
    print(result)
else:
    print(result.to_markdown(index=False, numalign="left", stralign="left"))
```

# Visualization 1: Trust Score vs. Raw Rating

```python
# Assuming the DataFrame 'df' is available from the previous block

# Prepare the data for plotting and apply log transformation to Votes
df_plot = df[['Rating', 'Votes', 'Trust_Score']].copy()
df_plot['Log_Votes'] = np.log1p(df_plot['Votes'])

# Set up the plot
fig, ax = plt.subplots(figsize=(10, 6))

scatter = sns.scatterplot(
    x='Rating',
    y='Log_Votes',
    hue='Trust_Score',
    data=df_plot,
    palette='viridis',
    size='Trust_Score',
    sizes=(20, 200),
    alpha=0.7,
    ax=ax
)

ax.set_title('Restaurant Reliability: Trust Score vs. Raw Rating and Votes')
ax.set_xlabel('Raw Rating')
ax.set_ylabel('Log(Votes + 1) [Proxy for Popularity]')

# Add a colorbar
norm = plt.Normalize(df_plot['Trust_Score'].min(), df_plot['Trust_Score'].max())
sm = plt.cm.ScalarMappable(cmap="viridis", norm=norm)
sm.set_array([])

fig.colorbar(sm, ax=ax, label="Trust Score (Weighted Rating)")

plt.tight_layout()
plt.savefig('trust_score_visualization.png')
print("Trust Score Visualization saved as trust_score_visualization.png")
```

# Visualization 2: Health Feature Distribution

```python
# Assuming the DataFrame 'df' is available from the previous block

# 1. Categorize Health Score
def categorize_health(score):
    if score >= 1:
        return 'Healthy Focus'
    elif score <= -1:
        return 'Indulgent Focus'
    else:
        return 'Neutral / Mixed'

df['Health_Category'] = df['Health_Score'].apply(categorize_health)

# 2. Count the occurrences of each category
health_counts = df['Health_Category'].value_counts().reset_index()
health_counts.columns = ['Health_Category', 'Count']

# 3. Plotting the bar chart
plt.figure(figsize=(8, 5))
sns.barplot(
    x='Health_Category',
    y='Count',
    data=health_counts,
    palette=['#4daf4a', '#377eb8', '#e41a1c']
)

plt.title('Distribution of Engineered Health Categories')
plt.xlabel('Health Focus (Based on Cuisine Tags)')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('health_category_distribution.png')
print("Health Category Distribution Visualization saved as health_category_distribution.png")
```
